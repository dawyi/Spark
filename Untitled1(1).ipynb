```python
dataFile = sc.textFile("/home/amir/data/present/timeSerie/calls.csv")
header = dataFile.first()
print(header)

data = dataFile.filter(lambda row: row != header).filter(lambda row: '1980' not in row)
print(data.first())
```

پارس کردن ورودی

```python
import time
import datetime

def callMap(row):
    (user,x,x,duration,datetime_) = row.split(',')
    dt = datetime.datetime.strptime(datetime_, "%a %b %d %H:%M:%S %z %Y")
    startDate = datetime.datetime.strptime('Sep 2 2009', "%b %d %Y")
    day = dt.date() - startDate.date()
    return (user, duration, dt.date(), datetime.datetime.strftime(dt, "%H"))

data = data.map(callMap)
print(data.first())
```

بررسی اولیه کاربر ها

```python
nu = data.map(lambda row: (row[0], 1)).reduceByKey(lambda a, b: a+b)
print ('number of users : ' + str(nu.count()))
for x in nu.collect(): print(x)
```

تعداد تماس های هر روز

```python
import matplotlib.pyplot as plt
import datetime
import numpy as np


cpd = data.map(lambda row: (row[2], 1)).reduceByKey(lambda a, b: a+b).sortByKey()

print(cpd.take(3))

x = np.array([i[0] for i in cpd.collect()])
y = np.array([i[1] for i in cpd.collect()])

plt.figure(figsize=(26,4))
plt.plot(x,y)
plt.show()
```

تعداد تماس ها در ساعت

```python
import matplotlib.pyplot as plt
import datetime
import numpy as np


cph = data.map(lambda row: (row[3], 1)).reduceByKey(lambda a, b: a+b).sortByKey()

print(cph.take(3))

x = np.array([i[0] for i in cph.collect()])
y = np.array([i[1] for i in cph.collect()])

plt.plot(x,y)
plt.show()
```

پیشبینی تعداد تماس ها در یک روز با استفاده از رگرسیون خطی 

(خط نارنجی نماینگر مقادیر پیشبینی شده است)


```python
from pyspark.ml.regression import LinearRegression
from pyspark.ml.linalg import Vectors


def regMap(row):
    return (float(row[1]) \
        ,Vectors.dense([float(datetime.datetime.strftime(row[0], "%y")) \
        ,float(datetime.datetime.strftime(row[0], "%m")) \
        ,float(datetime.datetime.strftime(row[0], "%d")) \
        ,1+float(datetime.datetime.strftime(row[0], "%w"))]))

regData = cpd.map(regMap)
regData = regData.toDF(['label','features'])

lr = LinearRegression()
model = lr.fit(regData)

predictions = model.transform(regData)

x = np.array([i[0] for i in cpd.collect()])
y = np.array([(i[0],i[2]) for i in predictions.collect()])

plt.figure(figsize=(26,4))
plt.plot(x,y)
plt.show()
    
```
